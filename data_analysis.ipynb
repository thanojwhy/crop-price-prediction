```python
import pandas as pd
```

```python
df=pd.read_csv('potato.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
df.head()
```

```python
df.to_csv("potato.csv",index=False)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x=df['Price'])
plt.axvline(df['Price'].mean(),linestyle="dashed",color='red')
plt.axvline(df['Price'].median(),linestyle="dashed",color='black')
```

```python
plt.boxplot(df['Price'],vert=False)
```

```python
ax=plt.figure(figsize=(8,6))
years=pd.DatetimeIndex(df['Date']).year
plt.scatter(x=df['Price'],y=years)
plt.title("Prices over the years of potato in India")
plt.ylabel('Years')
plt.xlabel('Price')
```

```python
df['Centre_Name'].unique()
```

```python
fig=plt.figure(figsize=(9,6))
plt.bar(x=df['Centre_Name'],height=df['Price'])
```

```python
df_blr=df[df['Centre_Name']=='BENGALURU']
```
