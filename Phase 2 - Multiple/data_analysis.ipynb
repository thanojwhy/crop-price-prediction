```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("all_commodities.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr(numeric_only=True)
```

```python
df.groupby(['State'])['State'].count()
```

```python
df=df[df['State']=='Karnataka']
```

```python
for crop in df['Commodity'].unique():
    prices=df[df.Commodity==crop]['Modal_Price']
    ax=plt.figure()
    plt.hist(x=prices)
    plt.axvline(prices.mean(),linestyle='dashed',color='red')
    plt.title(f"Price Distribution for {crop}")
    
```

```python
for crop in df['Commodity'].unique():
    prices=df[df.Commodity==crop]['Modal_Price']
    ax=plt.figure()
    plt.boxplot(x=prices,vert=False)
    plt.title(f"Price Statistics for {crop}")
```

```python
for crop in df['Commodity'].unique():
    prices=df[df.Commodity==crop][['Modal_Price','Year']]
    ax=plt.figure()
    plt.scatter(x=prices['Modal_Price'],y=prices['Year'])
    plt.xlabel('Price per quintal')
    plt.ylabel('Year')
    plt.title(f"Prices vs Year for {crop}")
```
