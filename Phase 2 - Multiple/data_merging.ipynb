```python
import pandas as pd
```

```python
prices=pd.read_csv('prices.csv')
yld=pd.read_csv('yield.csv')
```

```python
prices.drop(columns="Unnamed: 0",inplace=True)
yld.drop(columns="Unnamed: 0",inplace=True)
```

```python
prices.head()
```

```python
yld.head()
```

```python
prices['Year']=pd.to_datetime(prices['Arrival_Date'],format='mixed').dt.year
```

```python
yld=yld.rename(columns={'Crop_Year':'Year','Crop':'Commodity'})
yld.columns
```

```python
final=prices.merge(yld,how='left',on=['State','Commodity','Year'])
final.head()
```

```python
final['State'].unique()
```

```python
final['Commodity'].unique()
```

```python
final['Commodity_Code'].unique()
```

```python
final['Year'].unique()
```

```python
final=final.sort_values(by=['Year','State']).reset_index(drop=True)
```

```python
final.to_csv('all_commodities.csv')
```
